# **Analysis On Air Quality**



##**Problem Statement**

Air pollution has become a major environmental and public health concern across cities. Monitoring air quality is essential to identify polluted regions, understand pollutant trends, and support decision-making for environmental policies.

The dataset contains air quality measurements collected from multiple monitoring stations across different cities and states. It includes information such as:

* Country, state, and city
* Monitoring station location (latitude & longitude)
* Pollutant type (e.g., PM2.5, PM10, NO₂, etc.)
* Minimum, maximum, and average pollutant levels
* Last updated timestamp

##**Objective**

The goal of this project is to analyze air quality data to:

Identify cities and states with high pollution levels
Compare pollutant concentration across regions
Detect trends and variations in pollutant measurements
Support data-driven insights for environmental awareness and planning

##**Outcome**

The project provides insights into pollution distribution, highlights highly affected cities, and visualizes pollutant trends to support environmental monitoring and decision-making.

##**Domain**

The domain of the project is Environmental Science

##**Dataset Information**

Source: Government Portals/ data.gov.in/ image.png

Year / Timeline: Data collected during 2026

Dataset Column/features Description:


* Country: Name of the country where the monitoring station is located
* State: State or administrative region where the monitoring station is located
* City: City in which the air quality monitoring station is installed.
* Station: Name of the air quality monitoring station.
* Last update: Timestamp indicating when the air quality data was last recorded or updated.
* Latitude: Geographic latitude of the monitoring station.
* Longitude: Geographic longitude of the monitoring station.
* Pollutant Id: Indicates the type of pollutant measured.
* Pollutant Min: Minimum recorded value of the pollutant during the observation period.
* Pollutent Max: Maximum recorded value of the pollutant during the observation period.
* Pollutent Avg: Average concentration of the pollutant during the observation period.


###**Types Of Analysis To Be Done**



*   To summarize and understand the current state of air pollution in the dataset.
*   To identify causes and patterns behind high or low pollution levels.
*  To forecast future air quality trends using historical data.
*  To recommend actions to reduce pollution based on analysis insights.

# Identifying outliers using Box visualisation

# Identifying outliers using IOR method

```python
Q1 = df['pollutant_avg'].quantile(0.25)
Q3 = df['pollutant_avg'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['pollutant_avg'] < lower_bound) | (df['pollutant_avg'] > upper_bound)]
outliers.shape
```

**Outlier Treatment is NOT Necessary**


* As we are performing an environmental analysis
* As the outliers may represent real-world pollution spikes
* As we are identifying high pollution event and reporting real conditions (e.g., worst air quality days)  
*  A very high PM2.5 value could indicate: Industrial pollution,
Traffic congestion, Festival/firecracker events

## Histogram Chart - Pollution value distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['pollutant_avg'], kde=True)

plt.title("Histogram of Pollutant Average")
plt.xlabel("Pollutant Average")
plt.ylabel("Frequency")

plt.show()
```

Histogram shows the distribution of pollutant values

**Features used**:Pollutant Average

**Interpretation**:The histogram of pollutant average values shows a right-skewed distribution, indicating that most observations fall within lower to moderate pollution levels, while a few extreme values represent high pollution events. This suggests that although air quality is generally stable, occasional spikes occur, likely due to environmental or human activities.

## Box Plot- Pollutant Average

```python
sns.boxplot(x=df['pollutant_avg'])

plt.title("Box Plot of Pollutant Average")
plt.xlabel("Pollutant Average")

plt.show()
```

A box plot displays:

Median (middle value)
Quartiles (Q1, Q3)
Spread of data (IQR)
Outliers (points outside whiskers)

Features Used: Pollutant average
Interpretation:The box plot of pollutant values shows a wide distribution with several outliers above the upper whisker, indicating the presence of high pollution spikes. The median lies closer to the lower quartile, suggesting a right-skewed distribution where most values are low to moderate, but a few extreme values significantly increase the range. These outliers likely represent real environmental conditions rather than errors.

# Bar Plot- Average Pollution by City

```python
# Top 10 cities based on average pollution
top_cities = df.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False).head(10)

sns.barplot(x=top_cities.index, y=top_cities.values)

plt.title("Top 10 Cities by Average Pollution")
plt.xlabel("City")
plt.ylabel("Average Pollution")
plt.xticks(rotation=45)

plt.show()
```

A bar plot helps to:

Compare values across categories
Identify highest and lowest contributors
Easily spot trends and differences

**Features Used**: cities and pollutant average

**Interpretation**:The bar plot shows the comparison of average pollution levels across cities and pollutant types. It highlights that certain cities have significantly higher pollution levels than others, indicating regional variation in air quality. Additionally, some pollutants contribute more heavily to overall pollution, suggesting the need for targeted environmental control measures.

# Correlation Heatmap Visualization

```python
corr = df[['pollutant_min','pollutant_max','pollutant_avg','latitude','longitude']].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap of Air Quality Data")

plt.show()
```

Correlation heat map shows relationship between numeric variables.

**Features Used**: Numeric columns

**Interpretation**: The correlation heatmap shows a strong positive relationship between pollutant minimum, maximum, and average values, indicating consistency in pollution measurements. Geographic variables such as latitude and longitude show weak correlation with pollutant levels, suggesting that pollution is influenced by multiple environmental and human factors rather than location alone.
